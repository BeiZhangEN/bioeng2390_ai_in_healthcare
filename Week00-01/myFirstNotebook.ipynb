```python
# Generate a csv with tabular data containing age, height and weight and gender of 20 students 
import pandas as pd

# Generate random data
import random
import numpy as np

np.random.seed(0)
random.seed(0)

TotalRecords = 20

# Generate random ages between 18 and 25
ages = np.random.randint(18, 26, TotalRecords)

# Generate random heights between 150 and 200
heights = np.random.randint(150, 201, TotalRecords)

# Generate random weights between 50 and 100
weights = np.random.randint(50, 101, TotalRecords)

# Generate TotalRecords count of either M or F for male or female in a np array
gender = np.random.choice(['M', 'F'], TotalRecords)

# Create a DataFrame
df = pd.DataFrame({'Age': ages, 'Height': heights, 'Weight': weights, 'Gender': gender})

# Save the DataFrame to a csv file
df.to_csv('myFile.csv', index=False)

df.head()
```
